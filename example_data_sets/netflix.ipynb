```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load data sets of netflix
netflix_df = pd.read_csv('./data_sets/netflix_data.csv')
netflix_df.head()
```

```python
# filter the data to remove TV shows and store as netflix_subset

netflix_subset =  netflix_df[netflix_df["type"] == "Movie"]

netflix_subset.head()
    
```

```python
# Investigate and subset the Netflix movie data, keeping only the columns "title", "country", "genre", "release_year", "duration", and saving this into a new DataFrame called netflix_movies.
netflix_movies = netflix_df[["title", "country", "genre", "release_year", "duration"]]
netflix_movies.head()
```

```python
# Filter netflix_movies to find the movies that are strictly shorter than 60 minutes, saving the resulting DataFrame as short_movies;
# inspect the result to find possible contributing factors.

short_movies = netflix_movies[netflix_movies["duration"] < 60]
short_movies.head(10)
```

```python
# Using a for loop and if/elif statements, iterate through the rows of
movies = []

for lable, row in netflix_movies.iterrows():
    if row['genre'] == 'Documentaries':
        movies.append('yellow')
    elif row['genre'] == 'Dramas':
        movies.append('green')
    elif row['genre'] == 'Crime TV':
        movies.append('red')
    else :
        movies.append("nothing")

print(movies)
```

```python
plt.figure(figsize=(12, 8))

plt.scatter(netflix_movies["release_year"], netflix_movies['duration'], color = 'red')

plt.title("Movie Duration by Years of Release")
plt.xlabel("Release Year")
plt.ylabel("Duration (min)")

plt.show()
```
